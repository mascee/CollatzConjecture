## Collatz Conjecture
## Start with any number. If it's even, divide by 2. If it's odd, multiply by 3 and add 1.

```python
import matplotlib.pyplot as plt
```

```python
def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        sequence.append(n)
    return sequence
```

```python
# 📌 Cell 3: Run a test input
n = int(input("Enter a starting number: "))
seq = collatz_sequence(n)
print(f"Collatz sequence for {n}:")
print(seq)
print(f"Total steps: {len(seq)-1}")


# 📌 Cell 4: Plot sequence values
plt.figure(figsize=(10, 4))
plt.plot(seq, marker='o')
plt.title(f"Collatz Sequence Starting from {n}")
plt.xlabel("Step")
plt.ylabel("Value")
plt.grid(True)
plt.show()

```

```python
# 📌 Cell 5: Plot steps to reach 1 for multiple starting numbers
def collatz_steps(n):
    steps = 0
    while n != 1:
        steps += 1
        n = n // 2 if n % 2 == 0 else 3 * n + 1
    return steps
```

```python
max_n = 100
x_vals = list(range(1, max_n + 1))
y_vals = [collatz_steps(i) for i in x_vals]

plt.figure(figsize=(12, 5))
plt.bar(x_vals, y_vals, color='skyblue')
plt.title(f"Starting number {n}. {collatz_steps(n)} Steps to Reach 1 (Collatz Conjecture)")
plt.xlabel("Starting Number")
plt.ylabel("Steps")
plt.show()
```
